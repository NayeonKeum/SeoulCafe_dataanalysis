```
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
hasnum_inheo=pd.read_csv('/content/drive/My Drive/Colab Notebooks/data1/hasnum_inheo.csv')
```

```
hasnum_inheo.shape
hasnum_inheo.fillna(0)
```

```
hasnum_inheo['평균면적']=0
n=0
total=0
#hasnum_inheo.replace(to_replace=np.nan,value=0)
```

```
hasnum_inheo.replace(np.nan, 0)
for i in range(0,3227):
  if i==3226:
    break
  else:
    n+=1##값이 너무 소숫점 버리는데...
    total+=hasnum_inheo['소재지면적'][i]
    if hasnum_inheo['상권_코드'][i]!=hasnum_inheo['상권_코드'][i+1]:
      if n==1:
        hasnum_inheo['평균면적'][i]=total
      else:
        hasnum_inheo['평균면적'][i]=total/n
      n=0
      total=0

    else:
      continue
      ##2 결측값을 영구적으로 없앨 수 없는건가??애들이 존나 병신이야
```
